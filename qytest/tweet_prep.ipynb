```python
import pandas as pd
import numpy as np
import os
import nltk

from utils import preprocess_text
```

```python
# df = pd.read_csv("tweets.csv", on_bad_lines='skip')
df = pd.read_csv("tweets.csv")
df = df.drop(['id'], axis=1)
```

```python
cleaned_text = []
for t in df['text']:
    cleaned_text.append(' '.join(preprocess_text(t)[:-1]))
```

```python
cleaned_text = pd.DataFrame(cleaned_text, columns=['cleaned_text'])   
```

```python
cleaned_df = pd.concat([df, cleaned_text], axis=1, join='inner')
```
